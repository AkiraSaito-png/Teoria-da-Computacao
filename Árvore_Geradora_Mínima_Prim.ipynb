```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")
G.add_node("E")
G.add_node("F")
G.add_edge("A","B", weight = 12.5 )
G.add_edge("A","C", weight = 8.3 )
G.add_edge("C","D", weight = 5.2 )
G.add_edge("C","E", weight = 9.7 )
G.add_edge("C","F", weight = 2.3 )
G.add_edge("A","D", weight = 7.4 )

T = nx.minimum_spanning_tree(G, weight='weight', algorithm='prim', ignore_nan=False)

print('custo da AGM: ', G.size(weight="weight"))

plt.figure(1), nx.draw(G, with_labels=True, edge_color="b"), plt.title('grafo original: ')
plt.figure(2), nx.draw(T, with_labels=True, node_color="r", edge_color="r"), plt.title('AGM produzida pelo algoritmo: ')
plt.show
```
